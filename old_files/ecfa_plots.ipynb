```python
import ROOT
```

```python
ROOT.gROOT.LoadMacro("CLICdpStyle.C")
ROOT.CLICdpStyle()
```

```python
ROOT.gStyle.SetPadRightMargin(0.13)
ROOT.gStyle.SetPadLeftMargin(0.1)
```

```python
file = ROOT.TFile.Open("Output/final/resolutions.root", "read")
```

```python
mu_dir = file.mu_50deg_50GeV_1000evt
```

```python
mu_dir.ls()
```

```python
mu_pt = mu_dir.sdelta_pt.Clone()
e_pt = file.e_50deg_50GeV_1000evt.sdelta_pt.Clone()
```

```python
c_mu_pt = ROOT.TCanvas()
mu_pt.Draw()
c_mu_pt.Draw()
c_mu_pt.SaveAs("ecfa_mu_pt.pdf")

c_e_pt = ROOT.TCanvas()
e_pt.Draw()
c_e_pt.Draw()
c_e_pt.SaveAs("ecfa_e_pt.pdf")
```
