```python
df = ROOT.RDataFrame("events","reco_output/single_mu-_10GeV_40deg_lcio_rec_noforce_edm4hep.root")
```

```python
%%cpp
ROOT::VecOps::RVec<edm4hep::Vector3d> getTrackHitPos(ROOT::VecOps::RVec<edm4hep::TrackerHit> hit)
{
    ROOT::VecOps::RVec<edm4hep::Vector3d> res;

    for (size_t i = 0; i < hit.size(); i++) {
        res.push_back(hit[i].position);
    }
    return res;
}
```

```python
%%cpp
ROOT::VecOps::RVec<edm4hep::Vector3d> getSimTrackHitPos(ROOT::VecOps::RVec<edm4hep::SimTrackerHit> hit)
{
    ROOT::VecOps::RVec<edm4hep::Vector3d> res;

    for (size_t i = 0; i < hit.size(); i++) {
        res.push_back(hit[i].position);
    }
    return res;
}
```

```python
df2 = (df
       .Define("SimHit_pos", "getSimTrackHit(VertexBarrelCollection)")
       .Define("SimHit_R", "sqrt(SimHit_pos.x * SimHit_pos.x + SimHit_pos.y * SimHit_pos.y)")
       .Define("RecHit_pos", "getTrackHit(VXDTrackerHits)")
       .Define("RecHit_R", "sqrt(RecHit_pos.x * RecHit_pos.x + RecHit_pos.y * RecHit_pos.y)")
)
```

```python
h_sim_r = df2.Histo1D("SimHit_R")
h_rec_r = df2.Histo1D("RecHit_R")
```

```python
c_sim_r = ROOT.TCanvas()
h_sim_r.Draw()
c_sim_r.Draw()

c_rec_r = ROOT.TCanvas()
h_rec_r.Draw()
c_rec_r.Draw()
```
