```python
import ROOT
from math import floor, ceil
```

```python
%jsroot on
```

```python
df = ROOT.RDataFrame("events","Output/stage2/mu_89deg_5GeV_1000evt.root")
df.Describe()
```

```python
df2 = (df
       .Define("res_pt", "delta_pt / (true_pt * true_pt)")
)
```

```python
h_res_pt = df2.Histo1D("res_pt")
h_res_pt2 = df2.Histo1D(("res_pt", ";#Delta P_{T} / P_{T,true}^{2}", 50, 1, 0), "res_pt")
```

```python
g_res_pt2 = df2.Graph("res_pt", "rdfentry_")
count_rp = df2.Count()
v_res_pt2_rp = df2.Take["double"]("res_pt")
```

```python
c_res_pt = ROOT.TCanvas()
h_res_pt.Draw()
c_res_pt.Draw()

c_res_pt2 = ROOT.TCanvas()
h_res_pt2.Draw()
c_res_pt2.Draw()
```

```python
#count = count_rp.GetValue()
#g_res_pt2.Sort()
#g_res_pt2_low = g_res_pt2.GetPointX(ceil(0.01 * count))
#g_res_pt2_high = g_res_pt2.GetPointX(floor(0.99 * count))

# replace this by an std::sort on an std::vector??

#print(g_res_pt2_low)
#print(g_res_pt2_high)

v_res_pt2 = sorted(v_res_pt2_rp.GetValue())

g_res_pt2_low = v_res_pt2[ceil(0.01 * len(v_res_pt2))]
g_res_pt2_high = v_res_pt2[floor(0.99 * len(v_res_pt2))]

h_res_pt2_opt = df2.Filter(f"res_pt > {g_res_pt2_low} && res_pt < {g_res_pt2_high}").Histo1D(("res_pt", ";#Delta P_{T} / P_{T,true}^{2}", 50, g_res_pt2_low, g_res_pt2_high), "res_pt")
```

```python

ROOT.gStyle.SetOptFit(1111)
ROOT.gStyle.SetOptStat(110001111)
h_res_pt2_opt.Fit("gaus", "s")
```

```python
c_res_pt2_opt = ROOT.TCanvas()
h_res_pt2_opt.Draw()
c_res_pt2_opt.Draw()
h_res_pt2_opt.GetSkewness()
```
